```python
# Test 4: Check for source CSV files
print("\n" + "=" * 80)
print("TEST 4: Source Files Check")
print("=" * 80)

dfms = ["brown_shipley", "pershing", "castlebay", "wh_ireland"]
period = "2026-03"

for dfm_id in dfms:
    landing_path = f"Files/landing/period={period}/dfm={dfm_id}/source/"
    try:
        files = [f for f in mssparkutils.fs.ls(landing_path) if not f.isDir and f.name.lower().endswith('.csv')]
        if len(files) > 0:
            print(f"✓ {dfm_id}: Found {len(files)} CSV file(s)")
            for f in files:
                print(f"    - {f.name}")
        else:
            print(f"✗ {dfm_id}: No CSV files found")
    except:
        print(f"✗ {dfm_id}: Landing path not accessible")

print("\n[INFO] If CSV files are missing:")
print("  1. Upload your source CSV files to the appropriate dfm folder")
print("  2. Navigate to Files/landing/period=2026-03/dfm=<dfm_name>/source/")
print("  3. Click 'Upload' and select your CSV files")
```
